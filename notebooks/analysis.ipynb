# 🏦 Bank Customer Churn Analysis
This notebook contains the Exploratory Data Analysis (EDA) for the Bank Customer Churn project, focusing on identifying key drivers of churn and estimating revenue at risk.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set plotting style
sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# Load dataset
df = pd.read_csv('../data/bank_churn.csv')

# Drop useless columns
df.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1, inplace=True)

# Basic look
print("Dataset Shape:", df.shape)
print("\nMissing Values:")
print(df.isnull().sum())
print("\nClass Distribution (Exited):")
print(df['Exited'].value_counts())
```

```python
# Churn rate
churn_rate = df['Exited'].mean() * 100
print(f"Churn Rate: {churn_rate:.2f}%")

# Revenue at risk
revenue_at_risk = df[df['Exited'] == 1]['Balance'].sum()
print(f"Revenue at Risk: ${revenue_at_risk:,.2f}")
```

## Key Analysis Blocks

### Churn by Age Group

```python
df['AgeGroup'] = pd.cut(df['Age'], bins=[0,30,40,50,60,100],
                         labels=['<30','30-40','40-50','50-60','60+'])

churn_by_age = df.groupby('AgeGroup', observed=False)['Exited'].mean() * 100

plt.figure(figsize=(8,4))
sns.barplot(x=churn_by_age.index, y=churn_by_age.values, palette='Reds')
plt.title('Churn Rate by Age Group')
plt.ylabel('Churn Rate (%)')
plt.xlabel('Age Group')
plt.savefig('churn_by_age.png', dpi=300, bbox_inches='tight')
plt.show()
```

### Churn by Number of Products

```python
churn_by_products = df.groupby('NumOfProducts', observed=False)['Exited'].mean() * 100

plt.figure(figsize=(8,4))
sns.barplot(x=churn_by_products.index, y=churn_by_products.values, palette='Blues')
plt.title('Churn Rate by Number of Products')
plt.ylabel('Churn Rate (%)')
plt.xlabel('Number of Products')
plt.show()
```

### Churn by Geography

```python
churn_by_geo = df.groupby('Geography', observed=False)['Exited'].mean() * 100

plt.figure(figsize=(8,4))
sns.barplot(x=churn_by_geo.index, y=churn_by_geo.values, palette='Greens')
plt.title('Churn Rate by Country')
plt.ylabel('Churn Rate (%)')
plt.xlabel('Geography')
plt.show()
```

### Active vs Inactive Members

```python
churn_by_active = df.groupby('IsActiveMember', observed=False)['Exited'].mean() * 100
print("Churn by Active Status (0 = Inactive, 1 = Active):")
print(churn_by_active)
```
